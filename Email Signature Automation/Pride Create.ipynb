```python
import docx
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.shared import Pt
from docx.shared import Inches
import docx.enum.text
import os
import mammoth
```

```python
for f in os.listdir(r'c:\users\chung\desktop\HB_Email\HB_Email'):
    name, ext = os.path.splitext(f)
    if ext == '.txt' and name.endswith("STAND"):
        doc = open(r'c:\users\chung\desktop\HB_Email\HB_Email\{}'.format(f), encoding='cp1252')
        
        yourList = doc.readlines()
        new = []
        for ele in yourList:
            ele = ele.split('\n')
            ele = ''.join(ele)
            new.append(ele)          

        new_doc = docx.Document()
        
        pride = "Hanna Brophy Proudly Supports Pride Month and the LGBTQ+ Community"
        HB = "This is a privileged and confidential legal communication, protected by attorney/client privilege and is the exclusive property of the intended recipient of Hanna Brophy."
        URL = "www.hannabrophy.com"
        PO = 'P.O. Box 12488'
        address = 'Oakland, CA 94604-2488'
        #changing the page margins
        sections = new_doc.sections
        for section in sections:
            section.top_margin = Inches(1.0)
            section.bottom_margin = Inches(1.0)
            section.left_margin = Inches(1.0)
            section.right_margin = Inches(1.0)

        #setting style
        style = new_doc.styles['Normal']
        font = style.font
        paragraph_format = style.paragraph_format
        paragraph_format.space_after = Pt(0)
        paragraph_format.space_before = Pt(0)
        paragraph_format.line_spacing = 1
        font.name = 'Arial'
        font.size = Pt(12)

        #name bold font
        run1 = new_doc.add_paragraph().add_run(new[0])
        run1_font = run1.font
        run1_font.name = 'Arial'
        run1_font.size = Pt(12)
        run1_font.bold = True

        #title italicized
        run2 = new_doc.add_paragraph().add_run(new[1])
        run2_font = run2.font
        run2_font.name = 'Arial'
        run2_font.size = Pt(12)
        run2_font.italic = True

        new_doc.add_paragraph(new[2])
        new_doc.add_paragraph(new[3])
        new_doc.add_paragraph(new[4])
        
        #adding logo
        new_doc.add_picture(r"c:\users\chung\desktop\Output\picture1.png")
        
        #adding pride line
        new_doc.add_picture(r"c:\users\chung\desktop\Output\rainbow.png")

        #address
        run4 = new_doc.add_paragraph().add_run(URL)
        run4_font = run4.font
        run4_font.name = 'Arial'
        run4_font.size = Pt(12)

        
        run5 = new_doc.add_paragraph().add_run(PO)
        run5_font = run5.font
        run5_font.name = 'Arial'
        run5_font.size = Pt(12)

        
        run6 = new_doc.add_paragraph().add_run(address)
        run6_font = run6.font
        run6_font.name = 'Arial'
        run6_font.size = Pt(12)

        
        
        #last line
        run7 = new_doc.add_paragraph().add_run(HB)
        run7_font = run7.font
        run7_font.name = 'Arial'
        run7_font.size = Pt(8)
        run7_font.italic = True

        #save as docx
        new_doc.save(r'c:\users\chung\desktop\Output\{}.docx'.format(name))
        
        #save as rtf
        new_doc.save(r'c:\users\chung\desktop\Output\{}.rtf'.format(name))
        
        #save as text
        with open(r'c:\users\chung\desktop\Output\{}.txt'.format(name), "w") as text_file:
            for line in new[0:9]:
                print(line, file=text_file)
            print(pride, file=text_file)
            print(HB, file=text_file)
         
        #save as html
        html = "<p style='font-family: Arial' font-size='12px'><b>{}</b><br><em>{}</em><br>{}<br>{}<br><br><img src='picture1.png'><br><a href='www.hannabrophy.com'>{}</a><br>{}<br>{}<br><img src='rainbow.png' width='465px' height='20px'><br><em>{}</em></p>".format(
            new[0],
            new[1],
            new[2],
            new[3],
            URL,
            PO,
            address,
            HB
        )
        with open(r'c:\users\chung\desktop\Output\{}.html'.format(name), 'w') as html_file:
            html_file.write(html)
            
```
